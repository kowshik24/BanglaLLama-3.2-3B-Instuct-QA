```python
import requests

title_list = []
def get_recent_bengali_wikipedia_pages(limit=10):
    url = "https://bn.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "list": "recentchanges",
        "rcnamespace": 0,  # only get content pages, not talk pages, etc.
        "rclimit": limit,
        "rcprop": "title|timestamp",
        "format": "json"
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        recent_pages = data['query']['recentchanges']
        for page in recent_pages:
            print(f"Title: {page['title']}, Timestamp: {page['timestamp']}")
    else:
        print("Failed to fetch data from Wikipedia API")

get_recent_bengali_wikipedia_pages()
```

```python
import requests
title_list = []
def get_bangla_wikipedia_pageviews():
    url = "https://bn.wikipedia.org/w/api.php"

    # Set parameters to fetch the most viewed articles from Bengali Wikipedia
    params = {
        "action": "query",
        "format": "json",
        "list": "mostviewed",
        "pvimlimit": 1000,  # Fetch 10 most viewed pages
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        most_viewed_pages = data['query']['mostviewed']
        print("Most Viewed Bengali Wikipedia Articles (Weekly):")
        for page in most_viewed_pages:
            title = page['title']
            print(f"Title: {title}")
            title_list.append(title)
    else:
        print("Failed to retrieve data from Wikipedia API")

get_bangla_wikipedia_pageviews()
```

```python
import wikipediaapi
import pandas as pd

i = 0

# Initialize the Wikipedia API for Bengali language
wiki_wiki = wikipediaapi.Wikipedia('BanglaWiki', 'bn')

# Check if CSV file already exists
csv_file = 'Recent_Popular_Bangla_Wiki_Data.csv'

# Loop through each title in the list
for title in title_list:
    # Fetch the page
    page_py = wiki_wiki.page(title)

    # Check if the page exists
    if page_py.exists():
        # Prepare data for appending to CSV
        data = {
            'id': [i],  # You can adjust this ID as needed
            'title': [page_py.title],
            'text': [page_py.summary],
            'url': [page_py.fullurl]
        }
        i += 1

        # Create a DataFrame
        df = pd.DataFrame(data)

        # Append data to CSV file without overwriting existing data
        df.to_csv(csv_file, mode='a', index=False, header=not pd.io.common.file_exists(csv_file), encoding='utf-8-sig')

        print(f"Data added for title: {page_py.title}")
    else:
        print(f"Page does not exist for title: {title}")
```
